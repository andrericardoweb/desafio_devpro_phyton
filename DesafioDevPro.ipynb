Exercício 3 - Faça um Programa que peça dois números e imprima a soma.

```python
numero1 = eval(input('Digite um número: '))
numero2 = eval(input('Digite outro número: '))
soma = numero1 + numero2
print(f'{numero1} + {numero2} = {soma}')
```

Exercício 4 - Faça um Programa que peça as 4 notas bimestrais e mostre a média. 

```python
nota1 = eval(input('Nota 1: '))
nota2 = eval(input('Nota 2: '))
nota3 = eval(input('Nota 3: '))
nota4 = eval(input('Nota 4: '))
media = (nota1 + nota2 + nota3 + nota4) / 4
print(f'Média: {media:.2f}')
```

Exercício 5 - Faça um Programa que converta metros para centímetros.

```python
valor_metros = eval(input('Digite o valor em metros: '))
print(f'{valor_metros} metros é igual a {valor_metros * 100} centímetros')
```

Exercício 6 - Faça um Programa que peça o raio de um círculo, calcule e mostre sua área. 

```python
raio_circulo = eval(input('Digite o raio do círculo em metros: '))
pi = 3.14
area_circulo = pi * raio_circulo**2
print(f'A área do círculo é de {area_circulo}m²')
```

Exercício 7 - Faça um Programa que calcule a área de um quadrado, em seguida mostre o dobro desta área para o usuário.

```python
lado_quadrado = eval(input('Digite o tamanho de um lado do quadrado em metros:'))
area_quadrado = lado_quadrado**2
print(f'A área do quadrado são {area_quadrado}m² e o dobro são {area_quadrado * 2}m²')
```

Exercício 8 - Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês.

```python
ganho_por_hora = eval(input('Digite quanto você ganha por hora: R$'))
horas_trabalhadas_mes = eval(input('Digite quantas horas trabalhou no mês: '))
salario_mes = ganho_por_hora * horas_trabalhadas_mes
print(f'Total do salário no mês: R${salario_mes:.2f}')
```

Exercício 9 - Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius. C = 5 * ((F-32) / 9)

```python
temperatura_f = eval(input('Digite a temperatura em graus Fahrenheit: '))
converte_f_para_c = 5 * ((temperatura_f - 32) / 9)
print(f'{temperatura_f} graus Fahrenheit é igual a {converte_f_para_c:.1f} graus Celsius')
```

Exercício 10 - Faça um Programa que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.

```python
temperatura_celsius = eval(input('Digite a temperatura em graus Celsius: '))
converte_c_para_f = (temperatura_celsius * 1.8) + 32
print(f'{temperatura_celsius} graus Celsius é igual a {converte_c_para_f} graus Fahrenheit') 
```
